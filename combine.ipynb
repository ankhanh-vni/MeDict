```python
from glob import glob


"""
Get all paths to the files
"""
paths = glob(folder_path + "*.csv", recursive=True)
paths
```

```python
df = pd.read_csv(main_path)
df.head(2)
```

```python
df.shape
```

```python
to_be_concatted = []
for p in paths:
  df_append = pd.read_csv(p)
  to_be_concatted.append(df_append)
```

```python
final = pd.concat(to_be_concatted, ignore_index=True)
final.head()
```

```python
dropped = ['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13']
final = final.drop(dropped, axis=1)
final.shape
```

```python
final = pd.concat([df, final], ignore_index=True) 
final.head(2)
```

```python
final = final.drop('Unnamed: 0', axis=1)
final.shape
```
